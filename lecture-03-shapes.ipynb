# Shapes

Understand the shape of

$
F(q^2) = \frac{3[\sin(b) - b \cos(b)]}{b^3}
$

$
b = \frac{|q|a}{\hbar}
$

Where $q$ is momentum transfer (which basically goes as angle), and $a$ is the radius of the atom.

First, lets do theta - we'll plot from $-\pi/2$ to $\pi/2$.

```python
import numpy as np

theta_r = np.linspace(-np.pi/2, np.pi/2, 10000)
```

And to calculate $q$ from this we'll use the very simple relationship, assuming there is no recoil to first order.

![q2 calc](q2-calc.png)

```python
q_r = 2*(1-np.cos(theta_r))
```

```python
import matplotlib.pyplot as plt

plt.plot(theta_r, q_r)
plt.xlabel(r'$\theta$ [radians]')
plt.ylabel(r'$q$ with $|q| = 1$')
```

Since we aren't really interested in the behavior vs $a$, lets just make $b = q$.

```python
b_r = q_r*0.0000001
```

```python
import matplotlib.pyplot as plt

plt.plot(theta_r, b_r)
plt.xlabel(r'$\theta$ [radians]')
plt.ylabel(r'$b$')
```

And now we can calculate the form factor

```python
F_r = 3*(np.sin(b_r) - b_r*np.cos(b_r))/np.power(b_r, 3)
```

```python
import matplotlib.pyplot as plt

plt.plot(theta_r, F_r)
plt.xlabel(r'$\theta$ [radians]')
plt.ylabel(r'$F(q)$')
```
